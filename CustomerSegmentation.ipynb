```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read dataset
df= pd.read_csv('../Data/processed_data.csv')
# Convert date string to python date
Date= []
DateStr= [str(i[0:10]) for i in df.Date]
for d in DateStr:
    Date.append( datetime.datetime.strptime(d, "%Y-%m-%d").date() )
df= df.drop('Date',1)
df= df.drop('Unnamed: 0',1)
df['Date']= Date
df['WeekDay']= map(lambda x: x.weekday(), df.Date)
df.head()
```

```python
# Customer segmentation based on following features:
# mean_basket_recency, std_basket_recency, std_basket_value  <-- risk
# mean_basket_value, total_value, total_discounts  <-- profitability
```

Time to build aggregated customer dataset

```python
# start with aggregation per receipt
df2= df.groupby(['ReceiptNo'])
basket_recency= df2.Recency.agg(np.mean).reset_index()
basket_value= df2.Price.agg(np.sum).reset_index()
basket_discount= df2.apply(lambda x: np.sum( ( x.maxPrice - x.unitPrice ) * x.Amount ) ).reset_index()
```

```python
rec_df= df2.nth(0).reset_index(drop=True)
rec_df= rec_df.loc[:,['CustID','AgeCode','AreaCode','Recency','ReceiptNo','DistanceCode','Date','WeekDay']]
```

```python
buff= rec_df.copy()
rec_df= pd.merge( buff, basket_value, on="ReceiptNo", how='left' )
buff= rec_df.copy()
rec_df= pd.merge( buff, basket_discount, on="ReceiptNo", how='left' )
rec_df.columns= ['CustID','AgeCode','AreaCode','Recency','ReceiptNo',
                 'DistanceCode','Date','WeekDay','BasketValue','BasketDiscount']
rec_df.head()
```

```python
# Now aggregate per customer
df3= rec_df.groupby(['CustID'])
#
# Risk
mean_basket_recency= df3.Recency.agg(np.mean)
std_basket_recency= df3.Recency.agg(np.std)
std_basket_value= df3.BasketValue.agg(np.std)
#
# Profitability
mean_basket_value= df3.BasketValue.agg(np.mean)
total_value= df3.BasketValue.agg(np.sum)
total_discounts= df3.BasketDiscount.agg(np.sum)
```

```python
cust_df= df3.nth(0).reset_index()
cust_df= cust_df.ix[:,['CustID','AgeCode','AreaCode','DistanceCode']]
cust_df.head()
```

```python
buff= cust_df.copy()
cust_df= pd.merge( buff, mean_basket_recency.reset_index(), on="CustID", how='left' )
cust_df.columns.values[4]= 'MeanRecency'
#
buff= cust_df.copy()
cust_df= pd.merge( buff, std_basket_recency.reset_index(), on="CustID", how='left' )
cust_df.columns.values[5]= 'StdRecency'
#
buff= cust_df.copy()
cust_df= pd.merge( buff, std_basket_value.reset_index(), on="CustID", how='left' )
cust_df.columns.values[6]= 'StdBasketValue'
#
buff= cust_df.copy()
cust_df= pd.merge( buff, mean_basket_value.reset_index(), on="CustID", how='left' )
cust_df.columns.values[7]= 'MeanBasketValue'
#
buff= cust_df.copy()
cust_df= pd.merge( buff, total_value.reset_index(), on="CustID", how='left' )
cust_df.columns.values[8]= 'TotalValue'
#
buff= cust_df.copy()
cust_df= pd.merge( buff, total_discounts.reset_index(), on="CustID", how='left' )
cust_df.columns.values[9]= 'TotalDiscounts'
```

```python
NVisits= cust_df.TotalValue / cust_df.MeanBasketValue
cust_df.loc[:,'NVisits']= NVisits
```

```python
# fill NaN values in std fields with 'reasonable' numbers:
cust_df.StdRecency.fillna( 0.0,  inplace=True )
                          #cust_df.MeanRecency, inplace=True )
cust_df.StdBasketValue.fillna(  0.0 , inplace=True )
                              #cust_df.MeanBasketValue, inplace=True )
```

## Have a gander

```python
def scatterplot_matrix(df,refdf=pd.DataFrame(index=[],columns=[])):
    import matplotlib 
    from scipy.stats import gaussian_kde as spkd
    matplotlib.rc('xtick', labelsize=20) 
    matplotlib.rc('ytick', labelsize=20) 
    
    col= ['b','r','r','c','m']
    alpha= [0.2,0.2]

    plt.figure(1,figsize=(30,30))
    contlist= [#'CustID','AgeCode','AreaCode','DistanceCode',
               'MeanRecency','StdRecency','StdBasketValue',
               'MeanBasketValue','TotalValue','TotalDiscounts']#,'NVisits']
    for i in range(len(contlist)):
        nrows= len(contlist)-1
        adder=0.
        if not refdf.empty:
            nrows= len(contlist)
            ij= i*nrows+i+1
            sp= plt.subplot(nrows,nrows,ij)
            #
            kde= spkd( refdf.loc[:,contlist[i]] )
            okok= kde.evaluate(np.arange(200)/200.*np.max([ float(k) for k in refdf.loc[ :, contlist[i]]]))
            plt.plot(np.arange(200)/200.*np.max([ float(k) for k in refdf.loc[ :, contlist[i]]])
                     ,okok,'r')
            #
            kde2= spkd( df.loc[:,contlist[i]] )
            okok2= kde2.evaluate(np.arange(200)/200.*np.max([ float(k) for k in df.loc[ :, contlist[i]]]))
            plt.plot(np.arange(200)/200.*np.max([ float(k) for k in df.loc[ :, contlist[i]]])
                     ,okok2,'b')
            plt.xlabel( contlist[i],fontsize=20. )
            # ax.get_xaxis().set_visible(True)
            adder= 1.
        for j in range(i+1,len(contlist)):
            ij= i*nrows + j + adder
            sp= plt.subplot(nrows,nrows,ij)
            sp.scatter(df.loc[ :, contlist[j]],
                       df.loc[ :, contlist[i]],
                       c='b',alpha=0.1)
            plt.xlim([0.,np.max([ float(k) for k in df.loc[ :, contlist[j]]]) ])
            plt.ylim([0.,np.max([ float(k) for k in df.loc[ :, contlist[i]]]) ])
#            if j==i+1:
            plt.xlabel(contlist[j],fontsize=20.)
            plt.ylabel(contlist[i],fontsize=20.)
#             else:
#                 sp.get_xaxis().set_visible(False)
#                 sp.get_yaxis().set_visible(False)
    plt.tight_layout(pad=1.08)
    plt.show()
```
