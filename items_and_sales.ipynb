```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cluster
import sklearn.decomposition
import sklearn.preprocessing
import statsmodels
%matplotlib inline
```

```python
# Read dataset
df= pd.read_csv('../Data/processed_data.csv')
# Convert date string to python date
Date= []
DateStr= [str(i[0:10]) for i in df.Date]
for d in DateStr:
    Date.append( datetime.datetime.strptime(d, "%Y-%m-%d").date() )
df= df.drop('Date',1)
df= df.drop('Unnamed: 0',1)
df['Date']= Date
df['WeekDay']= map(lambda x: x.weekday(), df.Date)
df.head()
```

```python
# What items/classes etc are actually important?
```

```python
# aggregate by product:
dfgb= df.groupby('ProdID')
agg_price= dfgb.Price.agg(np.sum).reset_index()
agg_baskets= dfgb.apply(lambda x: len( x.ReceiptNo.unique() )).reset_index()
agg_customers= dfgb.apply(lambda x: len( x.CustID.unique() )).reset_index()
agg_items= dfgb.Amount.agg(np.sum).reset_index()
#
def count_sales(x):
    df= x.reset_index()
    sale= np.where(df.unitPrice <= 0.9 * df.maxPrice)[0]
    rtval= 0.
    if len(sale) > 0:
        rtval= np.sum(df.loc[sale,'Amount'])
    return rtval
#
agg_sale= dfgb.apply(count_sales).reset_index().fillna(0)
agg_sale.columns.values[1]= 'nDis'
```

```python
df2= pd.merge(  agg_price, df.loc[:,['ProdID','ProdClass','maxPrice']].drop_duplicates(),
              on='ProdID', how='left')
#
df2= pd.merge(  agg_baskets, df2,
              on='ProdID', how='left')
#
df2= pd.merge(  agg_customers, df2,
              on='ProdID', how='left')
#
df2= pd.merge(  agg_items, df2,
              on='ProdID', how='left')
#
df2= pd.merge(  agg_sale, df2,
              on='ProdID', how='left')
#
df2.columns.values[5]= 'Revenue'
df2.columns.values[4]= 'nBaskets'
df2.columns.values[3]= 'nCustomers'
df2.columns.values[2]= 'nSold'
#df2.columns.values[1]= 'nDiscounted'
```

```python
# aggregate by productClass:
#
df3= df.ix[:,['CustID','Amount','Price','maxPrice','unitPrice','ReceiptNo','ProdClass']]
perc= df3.unitPrice / df3.maxPrice
df3.loc[:,'Sale'] = map( lambda x: 1 if x <= 0.9 else 0, perc)
#
df3gb= df3.groupby('ProdClass')
#
rev= df3gb.Price.agg(np.sum).reset_index()
rev.columns.values[1]= 'Revenue'
#
baskets= df3gb.apply(lambda x: len( x.ReceiptNo.unique() )).reset_index()
baskets.columns.values[1]= 'nBaskets'
#
customers= df3gb.apply(lambda x: len( x.CustID.unique() )).reset_index()
customers.columns.values[1]= 'nCustomers'
#
discounted= df3gb.Sale.agg(np.sum).reset_index()
discounted.columns.values[1]= 'nDis'
#
```

```python
df_class= pd.merge( df3.drop_duplicates(), rev, on='ProdClass', how='left')
df_class= pd.merge( df_class, baskets, on='ProdClass', how='left')
df_class= pd.merge( df_class, customers, on='ProdClass', how='left')
df_class= pd.merge( df_class, discounted, on='ProdClass', how='left')
df_class.head()
```
